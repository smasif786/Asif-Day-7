```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import scikitlearn as
asi=pd.read_csv(r"C:\Users\HOME\Python\DataFiles\Climate.csv")
asi.head(2)
```

```python
asi =asi.drop(["Title"],axis=1,inplace=False)
asi.head(2)
```

```python
X = asi.iloc[:,1:]
X.astype(str)
X.head()
```

```python
y =asi.iloc[:,1]
y.head(2)
```

```python
# filling the missing value with most occuring value    
asi.iloc[:,:] = asi.iloc[:,:].apply(lambda x:x.fillna(x.value_counts().index[1]))
#claimants3.SEATBELT = claimants1.SEATBELT.fillna(claimants1.SEATBELT.value_counts().index[0])
asi.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
pred = reg.predict(X_test)
pred
```
