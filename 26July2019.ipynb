```python
import matplotlib.pyplot as plt
data_Asi={'Cat':100, "Lion" :150,"cow" :20, 'Horse':50, 'Goat':30}
names=list(data_Asi.keys())
values=list(data_Asi.values())
fig ,axs=plt.subplots(1,5,figsize=(10,5),sharey=True)
axs[0].bar(names,values)
axs[1].scatter(names,values)
axs[2].plot(names,values) 
fig.suptitle('Categerocal Plotting')
#axs[3].pie(names,values)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
division=['A','B','C','D']
avg_marks=[50,60,70,80]
marks=[76,67,55,46]
width=0.2
index=np.arange(len(division))
#plt.xlm(40.100)
plt.bar(index-width,marks,width=0.4,label='marks')
plt.bar(index+width,avg_marks,width=0.4,label='Avg_Marks')
plt.legend(loc='best') #
plt.xticks(index,division) #for conversion purpose
plt.show()
#plt.bar(avg_marks,division)
```

## pie chart

```python
lang=['python','java' ,'c', 'c++']
rating=[66,55,44,23]
plt.pie(rating,labels=lang.autopct='%1.%%f')
```

```python
lang=['python','java' ,'c', 'c++']
rating=[66,55,44,23]
plt.pie(rating,labels=lang,autopct='%1.1f%%', explode=[0,0,0.2,0],
        startangle=90,shadow=True,labeldistance=1.2)
```

```python
import matplotlib.pyplot as plotter

 

# The slice names of a population distribution pie chart

pieLabels  = 'Asia', 'Africa', 'Europe', 'North America', 'South America', 'Australia'

 

# Population data

populationShare     = [59.69, 16, 9.94, 7.79, 5.68, 0.54]

 

figureObject, axesObject = plotter.subplots()

 

# Draw the pie chart

axesObject.pie(populationShare,

        labels=pieLabels,

        autopct='%1.2f',

        startangle=90)

 

# Aspect ratio - equal means pie is a circle

axesObject.axis('equal')

 

plotter.show()
```
